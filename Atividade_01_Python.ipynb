```python
# Exercício 1
# Primeiro número
# float: converte para ponto flutuante (para aceitar decimais)
A = float(input("Digite o primeiro número: "))
# Segundo número
B = float(input("Digite o segundo número: "))

# Calcula a soma dos dois números e armazena o resultado na variável 'Soma'
Soma = A + B

# Exibe o resultado da soma
print("A soma é:", Soma)
```

```python
# Exercício 2
# Solicita a primeira nota
# float: converte para ponto flutuante (para aceitar decimais)
N1 = float(input("Digite a primeira nota: "))
# Solicita a segunda nota
N2 = float(input("Digite a segunda nota: "))

# Calcula a média aritmética das duas notas e armazena em 'Media'
Media = (N1 + N2) / 2

# Exibe o resultado da média na tela
print("A média é:", Media)
```

```python
# Exercício 3
# Primeiro número
# float: converte para ponto flutuante (para aceitar decimais)
X = float(input("Digite o primeiro número (X): "))
# Segundo número
Y = float(input("Digite o segundo número (Y): "))

# Verifica se o valor de X é estritamente maior que o valor de Y
if X > Y:
    # Se a condição for verdadeira, exibe que X é o maior
    print("X é maior")
# Caso contrário (se X não for maior ou se for igual a Y)
else:
    # Exibe que Y é o maior (ou igual, neste caso)
    print("Y é maior")
```

```python
# Exercício 4
# Solicita o número
# int: um número inteiro
N = int(input("Digite um número inteiro: "))

# O operador '%' (módulo) retorna o resto da divisão.
# Se o resto da divisão por 2 for 0, o número é Par.
if N % 2 == 0:
    print("O número é Par")
# Caso contrário, o número é Ímpar.
else:
    print("O número é Ímpar")
```

```python
# Exercício 5
# O contador começar em 1
contador = 1

# O loop 'while' continua executando enquanto o contador for menor ou igual a 10
while contador <= 10:
    # Calcula e imprime a multiplicação: contador x 5 = resultado
    print(contador, "x 5 =", contador * 5)

    # Passar para o próximo número (1, 2, 3...)
    contador += 1
```

```python
# Algoritmo 04 – Calcular a área de um triângulo

# 1. Solicita ao usuário os valores da base e da altura.
# float(): Permite números decimais (base e altura podem não ser inteiras).
base = float(input("Digite o valor da base do triângulo: "))
altura = float(input("Digite o valor da altura do triângulo: "))

# 2. Cacula área = (base * altura) / 2
area = (base * altura) / 2

# 3. Mostra o valor da área na tela
print("A área do triângulo é:", area)
```

```python
# Algoritmo 05 – Calcular a Área de um Triângulo (em Python)

# 1. Leia(b,h) - valores da base (b) e altura (h)
# float(): Permite números decimais
b = float(input("Digite o valor da base (b): "))
h = float(input("Digite o valor da altura (h): "))

# 2. área <- (b * h) / 2 - Calcula a área
area = (b * h) / 2

# 3. Mostre(área)
print("A área do triângulo é:", area)
```
